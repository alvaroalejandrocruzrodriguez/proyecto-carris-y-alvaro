```python
# Programa: Calculadora básica con condicionales
from fractions import Fraction

print("\nSeleccione la operación que desea realizar:")
print("1. Suma")
print("2. Resta")
print("3. Multiplicación")
print("4. División")

opcion = input("\nIngrese el número de la operación: ")


# Se piden los datos al usuario
a = Fraction(input("Digite el primer número: "))
b = Fraction(input("Digite el segundo número: "))


# Uso de condicionales
if opcion == "1":
    resultado = a + b
    print(f"La suma de {a} + {b} es: {resultado}")

elif opcion == "2":
    resultado = a - b
    print(f"La resta de {a} - {b} es: {resultado}")

elif opcion == "3":
    resultado = a * b
    print(f"La multiplicación de {a} * {b} es: {resultado}")

elif opcion == "4":
    if b != 0:
        resultado = a / b
        print(f"La división de {a} / {b} es: {resultado}")
    else:
        print("Error: no se puede dividir entre cero.")

else:
    print("Opción no válida. Por favor seleccione 1, 2, 3 o 4.")
```
